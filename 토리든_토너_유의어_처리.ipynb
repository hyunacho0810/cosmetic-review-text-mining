```python
import pandas as pd
from konlpy.tag import Okt
from gensim.models import Word2Vec

# 엑셀 파일에서 데이터 읽기
df = pd.read_excel('/content/토리든 토너 891개 (1).xlsx')
reviews = df['필드3'].tolist()

# KoNLPy 형태소 분석기 초기화
okt = Okt()

# 각 리뷰를 형태소 단위로 분할하여 리스트에 저장
tokenized_reviews = [okt.morphs(review) for review in reviews]

# Word2Vec 모델 학습
model = Word2Vec(sentences=tokenized_reviews, vector_size=100, window=5, min_count=5, sg=0)  # 벡터 사이즈와 하이퍼파라미터는 조정 가능

# 유의어 사전 구축
synonym_dict = {}

for word, _ in model.wv.most_similar('지성', topn=5):  # 원하는 단어에 대해 가장 유사한 단어 5개 추출
    if word not in synonym_dict:
        synonym_dict[word] = '원하는단어'  # 유사한 단어를 원하는 단어와 연결하여 유의어 사전 구축

# 전처리된 키워드를 유의어 사전을 활용하여 정규화
def normalize_keywords(keywords):
    normalized_keywords = []
    for keyword in keywords:
        if keyword in synonym_dict:
            normalized_keywords.append(synonym_dict[keyword])  # 유의어 사전에 있으면 대체하여 정규화
        else:
            normalized_keywords.append(keyword)
    return normalized_keywords

# 예시: 키워드 정규화 예시
preprocessed_keywords = ['맛집', '분위기', '맛있는', '음식']
normalized_keywords = normalize_keywords(preprocessed_keywords)
print("Normalized Keywords:", normalized_keywords)
```

```python
import nltk
nltk.download('wordnet')
from nltk.corpus import wordnet as wn

# nltk의 WordNet을 한국어로 설정 (Korean WordNet 사용)
nltk.download('omw-1.4', quiet=True)  # 한국어 WordNet 데이터 다운로드

def get_synonyms(word):
    synonyms = []
    for synset in wn.synsets(word, lang='kor'):
        for lemma in synset.lemmas(lang='kor'):
            synonym = lemma.name()
            if synonym != word:
                synonyms.append(synonym)
    return list(set(synonyms))  # 중복 제거를 위해 set으로 변환 후 다시 리스트로 변환

# 유의어 추출 예시
word = '사랑'
synonyms = get_synonyms(word)
print(f"유의어 '{word}': {synonyms}")
```

```python
from konlpy.corpus import kobill
from konlpy.tag import Kkma

# KoNLPy의 Kkma 형태소 분석기 초기화
kkma = Kkma()

# 한국어 텍스트 예시
text = "사랑은 가슴이 시리게 하는 감정입니다."

# 텍스트를 형태소로 분석하여 유의어 추출
morphs = kkma.morphs(text)
synonyms = []
for morph in morphs:
    if len(morph) > 1:  # 한 글자 형태소는 제외
        synsets = kkma.synsets(morph)
        for synset in synsets:
            synonyms.extend(synset.hypernyms())

# 중복 제거를 위해 set으로 변환 후 다시 리스트로 변환
synonyms = list(set(synonyms))

print(f"텍스트 '{text}'의 유의어: {synonyms}")
```

```python
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity
from konlpy.tag import Okt
import pandas as pd

# 엑셀 파일 경로 설정 (필요에 따라 경로를 수정하세요)
excel_file_path = '/content/토리든 토너 891개.xlsx'

# 엑셀 파일에서 데이터 불러오기 (필드3 열)
df = pd.read_excel(excel_file_path)
reviews = df['필드3'].tolist()  # 필드3 열의 데이터를 리스트로 변환

# 형태소 분석기 초기화 (Okt 사용)
okt = Okt()

# 모든 리뷰 데이터를 하나의 텍스트로 결합
combined_text = " ".join(reviews)

# 형태소 분석 및 명사 추출 (조사 제거)
nouns = [word for word, pos in okt.pos(combined_text) if pos == 'Noun']

# Word2Vec 모델 학습
model = Word2Vec(sentences=[nouns], vector_size=100, window=5, min_count=5, sg=0)

# 유사한 단어 추출 함수 정의
def get_similar_words(word, top_n=3):
    similar_words = []
    try:
        # 주어진 단어에 대해 유사한 단어 추출
        word_vector = model.wv[word]
        similar_word_vectors = model.wv.most_similar(word, topn=top_n)

        for similar_word, _ in similar_word_vectors:
            similar_words.append(similar_word)

    except KeyError:
        print(f"Word '{word}' not in vocabulary.")

    return similar_words

# 자주 등장하는 명사 선택 (30개 이하)
word_counts = pd.Series(nouns).value_counts()
top_nouns = word_counts.index.tolist()[:30]

# 각 명사에 대해 유사한 단어 추출하여 출력
similar_words_dict = {}

for noun in top_nouns:
    similar_words = get_similar_words(noun)
    similar_words_dict[noun] = similar_words

# 결과 출력
for noun, similar_words in similar_words_dict.items():
    print(f"명사 '{noun}'의 유의어: {similar_words}")
```

```python
!pip install PyKomoran  # PyKomoran 설치

import pandas as pd
from PyKomoran import Komoran
from google.colab import drive

# 입력 엑셀 파일 경로 설정 (필요에 따라 경로를 수정하세요)
input_excel_file_path = '/content/토리든 pykomoran을 이용한 전처리.xlsx'

# 출력 엑셀 파일 경로 설정 (필요에 따라 경로를 수정하세요)
output_excel_file_path = '/content/output_excel_file(1).xlsx'

# 엑셀 파일에서 데이터 불러오기 (필드3 열)
df = pd.read_excel(input_excel_file_path)
reviews = df['필드3'].tolist()  # 필드3 열의 데이터를 리스트로 변환

# PyKomoran을 사용하여 형태소 분석 및 Lemmatization 수행
komoran = Komoran("STABLE")  # 모델 경로를 지정하여 초기화 ('STABLE' 또는 'EXP')
w_ = []

def lemmatize(sentence):
    morphtags = komoran.pos(sentence)
    words = []
    for m, t in enumerate(morphtags) :
      k = t.get_pos()
      if k=='NNP' or k=='NNG' :
        words.append(t.get_morph())
      elif k=='VA' or k=='VV' :
        words.append(t.get_morph()+'다')
    return words

for review in reviews:
    try:
        words = lemmatize(review)
        w_.append(' '.join(words))
    except Exception as e:
        print(f"Error processing review: {review}")
        continue

# 데이터프레임에 단어 컬럼 추가
df['words'] = w_

# 빈 단어 제거
df = df[df['words'] != '']

# 결과를 새로운 엑셀 파일로 저장
df.to_excel(output_excel_file_path, index=False)

print(f"Data saved to {output_excel_file_path}")
```
