```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from collections import Counter
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 나눔고딕 폰트 설치
!apt-get update -qq
!apt-get install fonts-nanum* -qq

# 한글 폰트 설정
path = '/usr/share/fonts/truetype/nanum/NanumGothicEco.ttf'
fontprop = fm.FontProperties(fname=path, size=12)

# 데이터 불러오기 (preprocessed_review 열에 저장된 데이터를 사용한다고 가정)
from google.colab import files
uploaded = files.upload()

# 업로드한 파일명을 확인하고, 해당 파일을 DataFrame으로 읽어옵니다.
file_name = next(iter(uploaded))
df = pd.read_excel(io.BytesIO(uploaded[file_name]))

# TF-IDF 벡터화를 위한 객체 생성
tfidf_vectorizer = TfidfVectorizer()

# TF-IDF 행렬 생성
tfidf_matrix = tfidf_vectorizer.fit_transform(df['preprocessed_review'])

# 단어 목록 가져오기
words = tfidf_vectorizer.get_feature_names_out()

# 각 문서에서 TF-IDF 값이 가장 높은 단어 추출
top_words = []
for i in range(tfidf_matrix.shape[0]):
    # 문서 i의 TF-IDF 값과 단어 인덱스를 매핑
    word_scores = {words[idx]: score for idx, score in zip(tfidf_matrix[i].indices, tfidf_matrix[i].data)}
    # TF-IDF 값이 높은 순으로 정렬하여 상위 20개 단어 추출
    sorted_word_scores = sorted(word_scores.items(), key=lambda x: x[1], reverse=True)[:20]
    # 상위 단어 목록에 추가
    top_words.extend([word for word, score in sorted_word_scores])

# 상위 단어 중 가장 빈도가 높은 상위 20개 추출
top_korean_words = [word for word, _ in Counter(top_words).most_common(20)]

# 한국어로 출력
print("상위 20개 단어 (한국어):")
for idx, word in enumerate(top_korean_words):
    print(f"{idx+1}. {word}")

# 시각화: 단어 빈도 막대 그래프
word_counts = Counter(top_words)
word_df = pd.DataFrame(word_counts.items(), columns=['단어', '빈도']).sort_values(by='빈도', ascending=False)

plt.figure(figsize=(10, 8))
word_df[:20].plot(kind='barh', x='단어', y='빈도', ax=plt.gca(), color='skyblue')
plt.title('상위 20개 단어 빈도', fontproperties=fontprop)
plt.xlabel('빈도', fontproperties=fontprop)
plt.ylabel('단어', fontproperties=fontprop)
plt.gca().invert_yaxis()  # 빈도가 높은 순으로 표시
plt.savefig('word_frequency.png', bbox_inches='tight')  # 그래프를 이미지 파일로 저장
plt.show()
```
