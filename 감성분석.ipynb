```python
# 한국어 감성 사전 예시
korean_sentiment_dict = {
    '보습': 1,
    '수분': 1,
    '자극': -1,
    '건성': -1,
    '순하고': 1,
    '겨울': 0,
    '흡수': 0,
    '촉촉하고': 1,
    '무난': 0,
    '진정': 1,
    '화장': 0,
    '순해서': 1,
    '지성': -1,
    '트러블': -1,
    '여름': 0,
    '얼굴': 0,
    '복합': 0,
    '효과': 1,
    '있습니다': 0,
    '하게': 0,
    '발림': 0,
    '했는데': 0,
    '가격': 0,
    '아침': 0,
    '아주': 1,
    '바르기': 0,
    '건조': -1,
    '되는': 0,
    '완전': 1,
    '있어서': 0
}

# 단어별 감성 점수 계산 함수 정의
def calculate_word_sentiment(word, sentiment_dict):
    if word in sentiment_dict:
        return sentiment_dict[word]
    else:
        return 0  # 사전에 없는 단어는 중립적으로 처리

# 감성 분석을 수행할 단어 리스트
words_to_analyze = ["보습", "수분", "자극", "건성", "순하고","겨울","촉촉하고","무난","진정","화장","순해서","지성",
                    "트러블","여름","얼굴","복합","효과", "가격", "아침","건조"]

# 감성 분석 결과 출력
for word in words_to_analyze:
    sentiment_score = calculate_word_sentiment(word, korean_sentiment_dict)
    if sentiment_score > 0:
        sentiment_label = "긍정적"
    elif sentiment_score < 0:
        sentiment_label = "부정적"
    else:
        sentiment_label = "중립적"
    print(f"단어: {word}, 감성 점수: {sentiment_score}, 감성: {sentiment_label}")
```

```python
# 새로운 감성 사전 불러오기
def load_sentiment_dictionary(file_path):
    sentiment_dict = {}
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            word, score = line.strip().split('\t')
            sentiment_dict[word] = int(score)
    return sentiment_dict

# 새로운 감성 사전 파일 경로
new_sentiment_dict_file = "path/to/your/new/sentiment/dictionary.txt"  # 파일 경로를 적절하게 수정해주세요.

# 새로운 감성 사전 불러오기
new_korean_sentiment_dict = load_sentiment_dictionary(new_sentiment_dict_file)
```
