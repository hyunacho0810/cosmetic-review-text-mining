```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 텍스트 데이터 예시
text_data = "보습 수분 자극 건성 순하고 겨울 흡수 촉촉하고 진정 화장 순해서 지성 트러블 여름 얼굴 복합 효과 있습니다 하게 발림 했는데 가격 아침 아주 바르기 건조 되는 완전 있어서"

# 워드클라우드 생성 및 설정
wordcloud = WordCloud(font_path='NanumGothic.ttf',   # 한글 폰트 지정 (예: 나눔고딕)
                      width=800, height=400,          # 워드클라우드 이미지 사이즈
                      background_color='white',       # 배경색상
                      colormap='viridis').generate(text_data)

# 워드클라우드 시각화
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # 축 제거
plt.show()
```
