```python
import pandas as pd
import io
from konlpy.tag import Okt
from tqdm import tqdm

# KoNLPy의 Okt 형태소 분석기 객체 생성
okt = Okt()

# 데이터 불러오기
from google.colab import files

# 업로드한 파일을 DataFrame으로 읽어오기
uploaded = files.upload()

# 업로드한 파일을 BytesIO로 읽어오기
file_name = next(iter(uploaded))
file_content = uploaded[file_name]

# BytesIO로 읽은 파일을 DataFrame으로 변환
df = pd.read_excel(io.BytesIO(file_content))

# 조사 및 특정 단어 제거를 포함한 형태소 분석 함수 정의
def preprocess_and_tokenize(text):
    tokens = []
    previous_token = None

    for token, pos in okt.pos(text):
        # 조사 및 특정 단어 제거
        if pos != 'Josa' and token not in ['좋아요', '같아요', '이거', '진짜', '많이', '정말', '항상', '엄청', '계속', '너무', '것', '성',
                                           '좋은','있어요','쓰기','하는','재구매','느낌','바르고','피부','사용','제품','크림','수분크림','입니다','구매','좋고','없어','해서','쓰고','쓰는','없고','좋아서','했어요','하기',
                                           '않고','않아서','있는','되고','합니다','그냥','않고','닥터','없이','바르면','발라','없어서','라로슈포제','부지','타입','아직','생각','지금','추천',
                                           '확실히','자주','좋습니다','피지오']:
            # Compound word handling (e.g., preserve "데일리" as a single token)
            if previous_token is not None and previous_token in ['데']:
                # Merge current token with the previous one to form a compound word
                tokens[-1] += token
            else:
                tokens.append(token)

            previous_token = token

    return tokens

# 형태소 분석 및 조사 제거된 토큰 생성
tokenized_reviews = []
for review in tqdm(df['preprocessed_review']):
    tokens = preprocess_and_tokenize(review)
    tokenized_reviews.append(tokens)

# 결과를 DataFrame에 추가
df['tokenized_review'] = tokenized_reviews

# 결과 확인
print(df['tokenized_review'].head())

# 형태소 분석 및 조사 제거된 데이터를 엑셀 파일로 저장
output_file_name = 'preprocessed_reviews_tokenized.xlsx'
df.to_excel(output_file_name, index=False)
print(f"Preprocessed data saved to: {output_file_name}")
```
